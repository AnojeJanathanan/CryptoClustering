```python
# Imporpt required libraries and dependencies

import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

df_market_data.head(10)
```

```python
 
df_market_data.describe()
```

```python
 
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file

norm = pd.DataFrame(StandardScaler().fit_transform(df_market_data[["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d", "price_change_percentage_30d", "price_change_percentage_60d", "price_change_percentage_200d", "price_change_percentage_1y"]]), columns=["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d", "price_change_percentage_30d", "price_change_percentage_60d", "price_change_percentage_200d", "price_change_percentage_1y"])
norm
```

```python
# Create a DataFrame with the scaled data

normalization = pd.DataFrame(norm, columns=["price_change_percentage_24h", "price_change_percentage_7d","price_change_percentage_14d", "price_change_percentage_30d",
                                            "price_change_percentage_60d",  "price_change_percentage_200d", "price_change_percentage_1y"])

normalization["coin_id"] = df_market_data.index
normalization = normalization.set_index("coin_id")

print(normalization)
print(normalization.head())
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
val = list(range(1, 12))
val
```

```python
# Create an empty list to store the inertia values


# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

inertia = []
val = 11

for k in range(1, val+1):  
    cluster = KMeans(n_clusters=k, random_state=78)
    cluster.fit(norm)
    inertia.append(cluster.inertia_)

inertia

```

```python


dict_anoje = {"k": k, "inertia": inertia}
dict = pd.DataFrame(dict_anoje)
dict

```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
line_chart = dict.hvplot.line(
    x="k", 
    y="inertia",
) 

line_chart
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:**  The best val for k = 10 as shown above

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k

anoje = KMeans(n_clusters=10)
anoje
```

```python
# Fit the K-Means model using the scaled data
anoje.fit(normalization)
```

```python
 
crypto = anoje.predict(normalization)

crypto_df = pd.DataFrame(crypto)
crypto_df
```

```python
# Create a copy of the DataFrame
new = crypto_df.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
new['Predictions'] = crypto
 
new.head(13)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

plot = norm.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    c="Predictions",
    hover_cols=["coin_id"],
)

plot
```
